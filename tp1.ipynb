```python
import numpy as np
import matplotlib.pyplot as plt

# Defino funcion g(t)
def g(t):
    if t < -2 or t > 2:
        return 0
    elif -2 <= t < -1:
        return t + 1
    elif -1 <= t < 0:
        return 1
    elif 0 <= t < 1:
        return 2
    elif 1 <= t <= 2:
        return 2 - t

# rango de t
t_values_continuous = np.linspace(-3, 3, 400)
g_values_continuous = [g(t) for t in t_values_continuous]

# Plot g(t)
plt.figure(figsize=(8, 6))
plt.plot(t_values_continuous, g_values_continuous, label='$g(t)$', color='black', linewidth=2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-1, 3)
plt.title('$g(t)$')
plt.xlabel('$t$')
plt.ylabel('$g(t)$')
plt.grid(True)
plt.show()
```

```python
def f_par(func, x):
    return (func(x) + func(-x)) / 2

def f_impar(func,x):
    return (func(x) - func(-x)) / 2
```

### a) g(t) = 2t^3-3t+6

```python
import numpy as np
import matplotlib.pyplot as plt

def g(t):
    return 2*t**3 - 3*t + 6

t = np.linspace(0, 4, 100)
plt.plot(t, g(t), label='g(t) = 2t^3-3t+6', color='b', linestyle='-', linewidth=2)
plt.plot(t, f_par(g, t), label='g par', color='r', linestyle='--', linewidth=2)
plt.plot(t, f_impar(g, t), label='g impar', color='g', linestyle='--', linewidth=2)
plt.legend()
plt.grid(True)
plt.show()
```

### b) g(t) = 20 cos(40πt − π/4)

```python
import numpy as np
import matplotlib.pyplot as plt

def g(t):
    return 20 * np.cos(40 * np.pi * t - np.pi / 4)

t = np.linspace(0, 3 * np.pi, 100)
plt.plot(t, g(t), label='g(t)', color='b', linestyle='-.', linewidth=1)
plt.plot(t, f_par(g, t), label='g par', color='r', linestyle='-', linewidth=2)
plt.plot(t, f_impar(g, t), label='g impar', color='g', linestyle='-', linewidth=2)
plt.legend(loc='upper right')
plt.grid(True)
plt.xticks(np.arange(0, 3 * np.pi, np.pi), ['0', 'π', '2π'])
plt.xlabel('t')
plt.show()
```

### g[n] = e^(−(n/4)) u[n]

```python
# defino la funcion escalon u[n] discreta
def u(n):
    if n<0:
        return 0
    else: return 1
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u(n):
    return np.where(n >= 0, 1, 0)

def g(n):
    return np.exp(-(n/4)) * u(n)

def f_par(func, n):
    return (func(n) + func(-n)) / 2

def f_impar(func, n):
    return (func(n) - func(-n)) / 2

n = np.arange(-3, 5, 1)

# Plotear discreto con stem g(n), f_par(g(n)), f_impar(g(n))
plt.stem(n, g(n), label='g[n]', linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(n, f_par(g, n), label='g par[n]', linefmt='r-', markerfmt='rx', basefmt=' ')
plt.stem(n, f_impar(g, n), label='g impar[n]', linefmt='g-', markerfmt='g^', basefmt=' ')
plt.legend()
plt.grid(True)
plt.show()
```

### g[n] = sen[2πn/4] u[n]

```python
def g(n):
    return np.sin(n*2*np.pi/4) * u(n)

n = np.arange(-6, 11, 1)
plt.stem(n, g(n), label='g[n]', linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(n, f_par(g, n), label='g par[n]', linefmt='r-', markerfmt='rx', basefmt=' ')
plt.stem(n, f_impar(g, n), label='g impar[n]', linefmt='g-', markerfmt='g^', basefmt=' ')
plt.axvline(x=0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```
