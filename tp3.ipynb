```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

# Ejemplo 1

```python
t = sp.symbols('t')
k = sp.symbols('k')
# Def frec y periodo fundamental
f = 1
T = 1/f
# Calcula simbolicamnete la exepresion de los coeficientes de fourier
expk = sp.integrate(-1*sp.exp(-1*sp.I*k*2*sp.pi*f*t), (t, -1/2, 0))/T + sp.integrate(sp.exp(-1*sp.I*k*2*sp.pi*f*t), (t, 0, 1/2))/T
sp.pprint(expk)
```

```python
# Rango de los armonicos para los coeficientes
am = 25
armo = np.arange(-am, am+1, 1)
# Evaluacion de la expresion de los coeficientes de Fourier
coef = [complex(expk.subs(k, i).evalf()) for i in armo]
# Modulo de los coef de fourier
modcoef = np.abs(coef)
# Fase de los coef de fourier
fascoef = np.angle(coef)
# Grafica del modulo y fase de los coeficientes
plt.figure(figsize=(12, 6))

# Plot modulus of Fourier coefficients
plt.subplot(2, 1, 1)
plt.stem(armo, modcoef)
plt.title('Modulo de los Coeficientes de Fourier')
plt.xlabel('Armonicos')
plt.ylabel('Modulo')

# Plot phase of Fourier coefficients
plt.subplot(2, 1, 2)
plt.stem(armo, fascoef)
plt.title('Fase de los Coeficientes de Fourier')
plt.xlabel('Armonicos')
plt.ylabel('Fase')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
p = 1 # Ancho del escalon
t=sp.symbols('t')
k=sp.symbols('k')
T = 5 #Periodo
f = 1/T
# Calucle simbolicamente la expresion del coeficiente de fourier
expk = sp.integrate(sp.exp(-1*sp.I*k*2*sp.pi*f*t), (t, -p/2, p/2))/T
sp.pprint(expk)
```

```python
# Rango de los armonicos para los coeficientes
am = 25
armo = np.arange(-am, am+1, 1)
# Evaluacion de la expresion de los coeficientes de Fourier
coef = [complex(expk.subs(k, i).evalf()) for i in armo]
# Modulo de los coef de fourier
modcoef = np.abs(coef)
# Fase de los coef de fourier
fascoef = np.angle(coef)

# Grafica del modulo y fase de los coeficientes
plt.figure(figsize=(12, 6))

# Plot modulus of Fourier coefficients
plt.subplot(2, 1, 1)
plt.stem(armo, modcoef)
plt.title('Modulo de los Coeficientes de Fourier')
plt.xlabel('Armonicos')
plt.ylabel('Modulo')

# Plot phase of Fourier coefficients
plt.subplot(2, 1, 2)
plt.stem(armo, fascoef)
plt.title('Fase de los Coeficientes de Fourier')
plt.xlabel('Armonicos')
plt.ylabel('Fase')

plt.tight_layout()
plt.show()
```

## $b)$ 
Graficar los coeficientes de Fourier para la señal desplazada f(t-(p/2)). ¿Cómo se modifican dichos coeficientes? Tenga en cuenta que los coeficientes están en el dominio de la frecuencia.

```python
# Desplazo la señal en el dominio del tiempo aplicando propiedades de la serie de Fourier (multiplico por e^(-i*2*pi*k*f*p/2)
expk_2 = expk * sp.exp(-1*sp.I*k*2*sp.pi*f*(p/2))

# Evaluacion de la expresion de los coeficientes de Fourier desplazados
coef_2 = [complex(expk_2.subs(k, i).evalf()) for i in armo]
# Modulo de los coef de fourier desplazados
modcoef_2 = np.abs(coef_2)
# Fase de los coef de fourier desplazados
fascoef_2 = np.angle(coef_2)

# Grafica del modulo y fase de los coeficientes desplazados
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.stem(armo, modcoef_2)
plt.title('Modulo de los Coeficientes de Fourier Desplazados')
plt.xlabel('Armonicos')
plt.ylabel('Modulo')

plt.subplot(2, 1, 2)
plt.stem(armo, fascoef_2)
plt.title('Fase de los Coeficientes de Fourier Desplazados')
plt.xlabel('Armonicos')
plt.ylabel('Fase')

plt.tight_layout()
plt.show()
```

## $c)$
Graficar los coeficientes de Fourier para la señal comprimida f(2t). ¿Cómo se modifican dichos coeficientes? Tenga en cuenta que los coeficientes están en el dominio de la frecuencia.

```python
# Comprimir la señal en el dominio del tiempo aplicando propiedades de la serie de Fourier
a = 2
expk_compressed = expk * sp.exp(sp.I*k*a*2*sp.pi*f)

# Evaluacion de la expresion de los coeficientes de Fourier comprimidos
coef_compressed = [complex(expk_compressed.subs(k, i).evalf()) for i in armo]
# Modulo de los coef de fourier comprimidos
modcoef_compressed = np.abs(coef_compressed)
# Fase de los coef de fourier comprimidos
fascoef_compressed = np.angle(coef_compressed)

# Grafica del modulo y fase de los coeficientes comprimidos
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.stem(armo, modcoef_compressed)
plt.title('Modulo de los Coeficientes de Fourier Comprimidos')
plt.xlabel('Armonicos')
plt.ylabel('Modulo')

plt.subplot(2, 1, 2)
plt.stem(armo, fascoef_compressed)
plt.title('Fase de los Coeficientes de Fourier Comprimidos')
plt.xlabel('Armonicos')
plt.ylabel('Fase')

plt.tight_layout()
plt.show()
```
