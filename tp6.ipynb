# Actividad 1
$x(t) = 3\cdot \cos(100\cdot\pi\cdot t)$



```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2*0.01, 2*0.01, 0.0001)
x = 3 * np.cos(100 * np.pi * t)

# Frecuencias de muestreo
fm1 = 300
fm2 = 400
fm3 = 540

Tm1 = 1 / fm1
Tm2 = 1 / fm2
Tm3 = 1 / fm3

# Número de muestras
N1 = round(0.02 / Tm1)
N2 = round(0.02 / Tm2)
N3 = round(0.02 / Tm3)

# Índices de las muestras
n1 = np.arange(-N1, N1 + 1)
n2 = np.arange(-N2, N2 + 1)
n3 = np.arange(-N3, N3 + 1)

# Tiempos de muestreo
ts1 = n1 * Tm1
ts2 = n2 * Tm2
ts3 = n3 * Tm3

# Señales muestreadas
xs1 = 3 * np.cos(100 * np.pi * ts1)
xs2 = 3 * np.cos(100 * np.pi * ts2)
xs3 = 3 * np.cos(100 * np.pi * ts3)

# figura con subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 15))

# x(t)
axs[0].plot(t, x, label=r'$x(t) = 3 \cos(100 \pi t)$', color='blue')
axs[0].set_title(r'$x(t)$')
axs[0].set_xlabel(r'$t$ (s)')
axs[0].set_ylabel(r'$x(t)$')
axs[0].grid(True)
axs[0].legend()

# x[n] con T = 1/300 s
axs[1].plot(t, x, label=r'$x(t)$', color='blue', alpha=0.4)
axs[1].stem(ts1, xs1, label=r'$x[n]$ con $T = 1/300$ s', basefmt=" ", linefmt='r-', markerfmt='ro')
axs[1].set_title(r'$T = 1/300$ s')
axs[1].set_xlabel(r'$t$ (s)')
axs[1].set_ylabel(r'$x[n]$')
axs[1].grid(True)
axs[1].legend()

# x[n] con T = 1/400 s
axs[2].plot(t, x, label='$x(t)$', color='blue', alpha=0.4)
axs[2].stem(ts2, xs2, label=r'$x[n]$ con $T = 1/400$ s', basefmt=" ", linefmt='r-', markerfmt='ro')
axs[2].set_title(r'$T = 1/400$ s')
axs[2].set_xlabel(r'$t$ (s)')
axs[2].set_ylabel(r'$x[n]$')
axs[2].grid(True)
axs[2].legend()

# x[n] con T = 1/540 s
axs[3].plot(t, x, label=r'$x(t)$', color='blue', alpha=0.4)
axs[3].stem(ts3, xs3, label=r'$x[n]$ con $T = 1/540$ s', basefmt=" ", linefmt='r-', markerfmt='ro')
axs[3].set_title(r'$T = 1/540$ s')
axs[3].set_xlabel(r'$t$ (s)')
axs[3].set_ylabel(r'$x[n]$')
axs[3].grid(True)
axs[3].legend()

plt.tight_layout()
plt.show()
```

# Actividad 5

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
fm = 450  # Frecuencia de muestreo
bw = 150  # Ancho de banda del filtro pasa bajos
Tm = 1 / fm  # Período de muestreo
t = np.arange(0, 0.02, Tm)  # Vector de tiempo

# Señal original
x = np.cos(2 * np.pi * 100 * t) + np.cos(2 * np.pi * 220 * t)

# Señal muestreada
n = np.arange(len(t))
ts = n * Tm
xs = np.cos(2 * np.pi * 100 * ts) + np.cos(2 * np.pi * 220 * ts)

# Calcular la FFT de la señal muestreada
X = np.fft.fft(xs)
frequencies = np.fft.fftfreq(len(X), Tm)

# Filtrar las frecuencias mayores al ancho de banda del filtro pasa bajos
X_filtered = X.copy()
X_filtered[np.abs(frequencies) > bw] = 0

# Graficar el espectro de la señal muestreada
plt.figure(figsize=(12, 6))

# Espectro original
plt.subplot(2, 1, 1)
plt.stem(frequencies, np.abs(X))
plt.title('Espectro de la Señal Muestreada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.xlim(-300,300)
plt.grid(True)

# Espectro filtrado
plt.subplot(2, 1, 2)
plt.stem(frequencies, np.abs(X_filtered))
plt.title('Espectro de la Señal Filtrada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.xlim(-300,300)
plt.grid(True)

plt.tight_layout()
plt.show()
# TODO Falta que continuen los espectros mas alla de 200 Hz
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal triangular X(f)
def triangular(f, W):
    return np.maximum(1 - np.abs(f) / W, 0)

# Ancho de banda
W = 1000

# Frecuencias de muestreo
fm1 = 1500
fm2 = 2500

# Definir el rango de frecuencias
f = np.linspace(-3*W, 3*W, 1000)

# Calcular el espectro de la señal original
X_f = triangular(f, W)

# Calcular los espectros de las señales muestreadas
X_fm1 = triangular(f - fm1, W) + triangular(f + fm1, W)
X_fm2 = triangular(f - fm2, W) + triangular(f + fm2, W)

# Crear una figura con subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Graficar el espectro de la señal original
axs[0].plot(f, X_f, label=r'$X(f)$', color='blue')
axs[0].set_title(r'$X(f)$')
axs[0].set_xlabel('Frecuencia (Hz)')
axs[0].set_ylabel('Amplitud')
axs[0].grid(True)
axs[0].legend()

# Graficar el espectro de la señal muestreada con fm = 1500 Hz
#axs[1].plot(f, X_f, label=r'$X(f)$', color='blue', alpha=0.3)
axs[1].plot(f, X_fm1, label=r'$X(f)$ muestreada con $f_m = 1500$ Hz', color='red')
axs[1].set_title(r'Espectro de la señal muestreada con $f_m = 1500$ Hz')
axs[1].set_xlabel('Frecuencia (Hz)')
axs[1].set_ylabel('Amplitud')
axs[1].grid(True)
axs[1].legend()

# Graficar el espectro de la señal muestreada con fm = 2500 Hz
#axs[2].plot(f, X_f, label=r'$X(f)$', color='blue', alpha=0.3)
axs[2].plot(f, X_fm2, label=r'$X(f)$ muestreada con $f_m = 2500$ Hz', color='red')
axs[2].set_title(r'Espectro de la señal muestreada con $f_m = 2500$ Hz')
axs[2].set_xlabel('Frecuencia (Hz)')
axs[2].set_ylabel('Amplitud')
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la sonda
V_per_C = 10  # mV/°C
T_min = 0     # °C
T_max = 100   # °C

# Resolución deseada
resolucion = 1  # °C

# a. El rango de entrada en mV
V_min = T_min * V_per_C
V_max = T_max * V_per_C
print(f"Rango de entrada: {V_min} mV a {V_max} mV")

# b. El número de bits del CAD
N = (V_max - V_min) / resolucion
bits = int(np.ceil(np.log2(N)))
print(f"Número de bits del CAD: {bits} bits")

# c. El gráfico de las salidas del CAD y su error aplicando el ejemplo 3

# Definir la señal de entrada (temperatura en °C)
t = np.arange(0, 1.02, 0.02)
f = 1
signal = np.sin(2 * np.pi * f * t)

# El rango de la señal (Vmax-Vmin) se divide en N intervalos con un ancho
# delta = (Vmax-Vmin)/N
# Para este caso delta = |1-(-1)|/8=0.25
delta = 0.25

# El vector particion divide el rango de la señal de entrada en N intervalos
# y el vector posee N-1 elementos.
particion = np.arange(-1 + delta, 1, delta)

# El vector codigo1 divide el rango de la señal de entrada en N intervalos
# Asigna un valor de cuantificación a cada región definida en particion.
codigo1 = np.arange(-1, 1, delta)

# El vector codigo2 divide el rango de la señal de entrada en N intervalos
# diferentes al anterior, pero también contiene N-1 elementos.
# Asigna un valor de cuantificación a cada región definida con particion.
codigo2 = np.arange(-0.875, 0.875 + delta, delta)

# La cuantización es un proceso que digitaliza una señal analógica.
# El proceso mapea valores de muestra de la entrada dentro de los intervalos
# del rango a diferentes valores comunes.
# El mapeo de cuantificación requiere una partición y un código.
# Utilice la función digitize para mapear una señal de entrada a una señal
# cuantizada escalar.
index1 = np.digitize(signal, particion)
quants1 = codigo1[index1 - 1]

# Se calcula el error1 entre el valor de la señal y el valor de cuantización.
error1 = np.abs(signal - quants1)

index2 = np.digitize(signal, particion)
quants2 = codigo2[index2 - 1]

# Se calcula el error2 entre el valor de la señal y el valor de cuantización.
error2 = np.abs(signal - quants2)

# Se grafican los valores antes calculados.
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Gráfico 1
axs[0].plot(t, signal, 'x', label='Señal original')
axs[0].plot(t, quants1, '*', label='Cuantización 1')
axs[0].plot(t, error1, '+', label='Error 1')
axs[0].set_title('Cuantización 1')
axs[0].set_xlabel('Tiempo (s)')
axs[0].set_ylabel('Amplitud')
axs[0].axis([-0.1, 1.2, -1.5, 1.5])
axs[0].grid(True)
axs[0].legend()

# Gráfico 2
axs[1].plot(t, signal, 'x', label='Señal original')
axs[1].plot(t, quants2, '*', label='Cuantización 2')
axs[1].plot(t, error2, '+', label='Error 2')
axs[1].set_title('Cuantización 2')
axs[1].set_xlabel('Tiempo (s)')
axs[1].set_ylabel('Amplitud')
axs[1].axis([-0.1, 1.2, -1.5, 1.5])
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la sonda
V_per_C = 10  # mV/°C
T_min = 0     # °C
T_max = 100   # °C

# Resolución deseada
resolucion = 1  # °C

# a. El rango de entrada en mV
V_min = T_min * V_per_C
V_max = T_max * V_per_C
print(f"Rango de entrada: {V_min} mV a {V_max} mV")

# b. El número de bits del CAD
N = (V_max - V_min) / resolucion
bits = int(np.ceil(np.log2(N)))
print(f"Número de bits del CAD: {bits} bits")

# c. El gráfico de las salidas del CAD y su error aplicando el ejemplo 3

# Definir la señal de entrada (temperatura en °C)
t = np.arange(0, 1, 0.01)
temperatura = 50 * np.sin(2 * np.pi * t) + 50

# Convertir la señal de temperatura a mV
signal = temperatura * V_per_C / 1000

# Definir la partición y los códigos de cuantificación
delta = (V_max - V_min) / (2**bits)
particion = np.arange(V_min + delta, V_max, delta)
codigo1 = np.arange(V_min, V_max, delta)
codigo2 = np.arange(V_min + delta/2, V_max + delta/2, delta)

# Cuantizar la señal
index1 = np.digitize(signal, particion)
quants1 = codigo1[index1 - 1]
error1 = np.abs(signal - quants1)

index2 = np.digitize(signal, particion)
quants2 = codigo2[index2 - 1]
error2 = np.abs(signal - quants2)

# Graficar los resultados
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Gráfico 1
axs[0].plot(t, signal, 'x', label='Señal original')
axs[0].plot(t, quants1, '*', label='Cuantización 1')
axs[0].plot(t, error1, '+', label='Error 1')
axs[0].set_title('Cuantización 1')
axs[0].set_xlabel('Tiempo (s)')
axs[0].set_ylabel('Voltaje (mV)')
#axs[0].axis([-0.1, 1.2, -1.5, 1.5])
axs[0].grid(True)
axs[0].legend()

# Gráfico 2
axs[1].plot(t, signal, 'x', label='Señal original')
axs[1].plot(t, quants2, '*', label='Cuantización 2')
axs[1].plot(t, error2, '+', label='Error 2')
axs[1].set_title('Cuantización 2')
axs[1].set_xlabel('Tiempo (s)')
axs[1].set_ylabel('Voltaje (mV)')
#axs[1].ylim(0,120)
axs[1].axis([-0.1, 1, -1, 100])
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
temp_min = 0    # Temperatura mínima
temp_max = 100  # Temperatura máxima
sensibilidad = 10  # mV/°C

# Generación de señal sinusoidal de temperatura
t = np.linspace(0, 1, 100)
signal = temp_max/2 * np.sin(2*np.pi*t) + temp_max/2

# Conversión de temperatura a milivoltios
signal_mv = signal * sensibilidad

# Parámetros de cuantización
n_bits = 7
n_niveles = 2**n_bits
particion = np.linspace(signal_mv.min(), signal_mv.max(), n_niveles-1)
codigo1 = np.linspace(signal_mv.min(), signal_mv.max(), n_niveles)

# Cuantización
def cuantizar(señal, particion, codigo):
    index = np.searchsorted(particion, señal)
    quants = codigo[index]
    return index, quants


# Realizar cuantización
index1, quants1 = cuantizar(signal_mv, particion, codigo1)

# Cálculo de error
error1 = np.abs(signal_mv - quants1)

# Gráficos
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal_mv, 'x', label='Señal Original (mV)')
plt.plot(t, quants1, '*', label='Señal Cuantizada (mV)')
plt.plot(t, error1, '+', label='Error de Cuantización (mV)')
plt.title(f'Cuantización con {n_bits} bits')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje (mV)')
plt.legend()
plt.grid(True)

# Segundo subplot con diferentes valores de código
codigo2 = np.linspace(signal_mv.min()+50, signal_mv.max()-50, n_niveles)
index2, quants2 = cuantizar(signal_mv, particion, codigo2)
error2 = np.abs(signal_mv - quants2)

plt.subplot(2, 1, 2)
plt.plot(t, signal_mv, 'x', label='Señal Original (mV)')
plt.plot(t, quants2, '*', label='Señal Cuantizada (mV)')
plt.plot(t, error2, '+', label='Error de Cuantización (mV)')
plt.title(f'Cuantización con Código Alternativo')
plt.xlabel('Tiempo')
plt.ylabel('Voltaje (mV)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Imprimir información adicional
print(f"Rango de entrada: 0 - {signal_mv.max()} mV")
print(f"Número de bits: {n_bits}")
print(f"Número de niveles: {n_niveles}")
print(f"Resolución: {sensibilidad} mV/nivel")
```
